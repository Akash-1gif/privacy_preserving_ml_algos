```python
n_samples,n_features=X_train.shape
print(n_samples,n_features)
```

```python
bias=0
var1=1/n_samples
n_iters=1000
lr=0.01
# weights as scalars:
w_a1=0
```

```python
bias_counts=[bias]*n_samples
bias_counts=cs.encrypt(bias_counts)
```

```python
dot_prod=enc_X_train_cols*w_a1
dot_prod
```

```python
y_pred=dot_prod+bias_counts
```

```python
test=cs.decrypt(y_pred)
test
```

```python
def offset_for_subtraction(l):
    l=cs.decrypt(l)
    offset=10000
    m=[]
    for i in l:
        p=offset+i
        m.append(p)

    m1=cs.encrypt(m)
    return m1
```

```python
enc_y_train_nve=offset_for_subtraction(enc_y_train)
test2=cs.decrypt(enc_y_train_nve)
test2
```

```python
res=y_pred+enc_y_train_nve
```

```python
def offset_removal(l):
    l=cs.decrypt(l)
    offset=10000
    m=[]
    for i in l:
        p=i-offset
        m.append(p)

    m1=cs.encrypt(m)
    return m1
```

```python
res=offset_removal(res)
test3=cs.decrypt(res)
test3
```

```python
def res_sum_fun(l):
    p=cs.decrypt(l)
    summation=0
    for i in p:
        summation+=i
    return summation
```

```python
res_sum=res_sum_fun(res)
res_sum
```

#### updating bias

```python
db=res_sum*var1
db=db*lr
```
