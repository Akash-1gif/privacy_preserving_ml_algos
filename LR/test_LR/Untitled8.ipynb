```python
from phe import paillier
```

```python
public_key, private_key = paillier.generate_paillier_keypair()
```

```python
x=-10000
x1=public_key.encrypt(x)

y=1000
y1=public_key.encrypt(y)
```

```python
p=x1+y1
private_key.decrypt(p)
```
